```python
# import necessary libraries
import pandas as pd

# Load the data set
df = pd.read_csv("heart.csv")
```

```python
df
```

```python
# Preprocessing / Data cleaning
# 1. Empty cells or missing data
print(df.isnull().sum())
```

```python
# 2. Duplicates
# Finding and displaying all duplicated rows(considering all columns)
duplicated_rows = df[df.duplicated(keep=False)] # 'keep=False' returns how many times that row appears
print("Duplicated rows across the dataframe")
print(duplicated_rows)
```

```python
# dropping one of the duplicated row
df = df.drop_duplicates()
# Checking for duplicates
remaining_duplicates = df[df.duplicated(keep=False)]
print("Remaining duplicates:")
print(remaining_duplicates)
```

```python
# 3. Wrong format
# checking for data formats for all columns
print(df.dtypes)
df.info()
```

```python
# checking numeric columns for wrong data format
numeric_cols = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    inconsistent_count = df[col].isna().sum()
    print(f"Inconsistent numeric value : {inconsistent_count}")
```

```python
# 4. Wrong data
# Visualize wrong data using boxplot to spot outliers
import matplotlib.pylab as plt

plt.figure(figsize=(12,8))
df.boxplot(column=numeric_cols)
plt.title("Box plot for numeric columns")
plt.show()
```

```python
# Define a function to Identifying outliers in each column using IQR
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# Identify outliers in each nmeric column
for columns in numeric_cols:
    outliers = identify_outliers(df, columns)
    print(f"Outliers in {columns}")
    print(outliers)
```
