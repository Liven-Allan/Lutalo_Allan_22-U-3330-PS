```python
import pandas as pd
df = pd.read_csv('niceguys.csv')
# fill missing values in 'living_allowance' with the median of that column
# 1 : first convert 'living_allowance' to numeric removing commas and handling errors

df['living_allowance'] = pd.to_numeric(df['living_allowance'].str.replace(',', ''), errors='coerce')

# 2 : then populate missing values with median in 'living_allowance' column
df['living_allowance'].fillna(df['living_allowance'].median())
```

```python
# Checking for missing values
print(df.isnull().sum())
```

```python
# EDA
# Summary statistics
print(pd.describe())
```

```python
# Drop row where living_allowance has a value 50,000 and 1,000,000
df[~df['living_allowance'].isin([50000,1000000])]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# matplotlib.pylot : is a module in the Matplotlib library. It provides a MATLAB-like interface f
# or creating figures and plots. pyplot provides functions for creating basic types of plots such 
# as line plots, scatter plots, bar plots, histograms, and more

# Seaborn is a Python data visualization library based on matplotlib. It provides a high-level 
# interface for drawing attractive and informative statistical graphics. 

# Convert the 'do_you_have_aboyfriend/girlfriend' column to string type
df['do_you_have_aboyfriend/girlfriend'] = df['do_you_have_aboyfriend/girlfriend'].astype(str)

# EDA (Expolatory Data Analysis)
# Distributing students with age and whether they have girlfriend or boyfriend
plt.figure(figsize=(10, 6))
sns.countplot(x='age', hue='do_you_have_aboyfriend/girlfriend', data=df )
plt.title("Distribution of Students by Age and Having a Girlfriend/Boyfriend")
plt.show()
```

```python
# Gender analysis of those who have a boyfriend or girlfriend
plt.figure(figsize=(10,6))
sns.countplot(x='sex', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title("Gender Analysis of Having a Boyfriend/Girlfriend")
plt.show()
```

```python
# Analysis by eating habits and having a girlfriend or boyfriend
plt.figure(figsize=(10,6))
sns.countplot(x='How_many_time_do_eat_food',  hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Analysis by eating habits and having a girlfriend or boyfriend')
plt.show
```

```python
# Analysis by age and having a girlfriend or boyfriend
plt.figure(figsize=(10,6))
sns.countplot(x='age',  hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Analysis by age and having a girlfriend or boyfriend')
plt.show
```

```python
# Analysis by gender and having a girlfriend or boyfriend
plt.figure(figsize=(10,6))
sns.countplot(x='sex',  hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Analysis gender and having a girlfriend or boyfriend')
plt.show
```

```python
# Relationship between living allowance and having a girlfriend or boyfriend
plt.figure(figsize=(10,6))
sns.boxplot(x='do_you_have_aboyfriend/girlfriend', y='living_allowance', data=df)
plt.title('Living Allowance and Having a Girlfriend/Boyfriend')
plt.show
```

```python
# Analysis of those with a girlfriend or boyfriend and whether they have siblings
plt.figure(figsize=(10,6))
sns.countplot(x='number_siblings', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Siblings and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Careers and having a girlfriend or boyfriend
plt.figure(figsize=(10,6))
sns.countplot(x='Are_you_a', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Siblings and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Step 5: Correlation analysis
# correlation analysis is used for spotting patterns with in the dataset

# positive correlations results means that both variables increase in relation to each other
# negative correlation results means that as one variable decreases the other one increases.

# select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include='number')

# compute correlation matrix
correlation_matric = numeric_columns.corr()

# plot heatmap
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matric, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heat map')
plt.show()
```

```python
# Step 6: Decision making / insights of the data, data analysis conclusion
# 1: Age and having a girlfriend or boyfriend
age_gf_bf = df.groupby('age')['do_you_have_aboyfriend/girlfriend'].value_counts(normalize=True).unstack()
print(age_gf_bf)
```
