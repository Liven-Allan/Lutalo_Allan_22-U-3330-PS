```python
import pandas as pd

# Read the CSV data
df = pd.read_csv('niceguys.csv')

# Convert 'living_allowance' to numeric, removing any commas 
df['living_allowance'] = pd.to_numeric(df['living_allowance'].str.replace(',', ''), errors='coerce')

# Filll missing values, with the median of the 'living_allowance' column
df['living_allowance'] = df['living_allowance'].fillna(df['living_allowance'].median())
print(df.head(5))

```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop row where living allowance has a value 50,000 and 1,000,000
df[~df['living_allowance'].isin([50000,1000000])]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the 'do_you_have_aboyfriend/girlfriend' column to string type
df['do_you_have_aboyfriend/girlfriend'] = df['do_you_have_aboyfriend/girlfriend'].astype(str)


# Distribution of students by age and whether they have a girlfriend or boyfriend
plt.figure(figsize=(10, 6))
sns.countplot(x='age', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Distribution of Students by Age and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Gender analysis of those who have a boyfriend or girlfriend
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Gender Analysis of Having a Boyfriend/Girlfriend')
plt.show()
```

```python
# Analysis by eating habits, age, gender, and having a girlfriend or boyfriend
plt.figure(figsize=(10, 6))
sns.countplot(x='How_many_time_do_eat_food', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Eating Habits and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Relationship between living allowance and having a girlfriend or boyfriend
plt.figure(figsize=(10, 6))
sns.boxplot(x='do_you_have_aboyfriend/girlfriend', y='living_allowance', data=df)
plt.title('Living Allowance and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Analysis of those with a girlfriend or boyfriend and whether they have siblings
plt.figure(figsize=(10, 6))
sns.countplot(x='number_siblings', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Siblings and Having a Girlfriend/Boyfriend')
plt.show()
```

```python
# Careers and having a girlfriend or boyfriend
plt.figure(figsize=(10, 6))
sns.countplot(x='Are_you_a', hue='do_you_have_aboyfriend/girlfriend', data=df)
plt.title('Careers and Having a Girlfriend/Boyfriend')
plt.xticks(rotation=90)
plt.show()
```

```python
# Step 5: Correlation analysis

# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include='number')

# Compute correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 6: Decision making / insights of the data, data analysis conclusion
# 1: Age and having a girlfriend or boyfriend
age_gf_bf = df.groupby('age')['do_you_have_aboyfriend/girlfriend'].value_counts(normalize=True).unstack()
print(age_gf_bf)
```

```python
# 2: Gender and having a girlfriend or boyfriend
gender_gf_bf = df.groupby('sex')['do_you_have_aboyfriend/girlfriend'].value_counts(normalize=True).unstack()
print(gender_gf_bf)
```

```python
# 3: Correlation analysis
print(correlation_matrix)
```

```python
# 4: Living allowance and having a girlfriend or boyfriend
living_allowance_gf_bf = df.groupby('do_you_have_aboyfriend/girlfriend')['living_allowance'].describe()
print(living_allowance_gf_bf)
```

```python
# 5: Careers and having a girlfriend or boyfriend
career_gf_bf = df.groupby('Are_you_a')['do_you_have_aboyfriend/girlfriend'].value_counts(normalize=True).unstack()
print(career_gf_bf)
```

```python
# 6: Siblings and having a girlfriend or boyfriend
siblings_gf_bf = df.groupby('number_siblings')['do_you_have_aboyfriend/girlfriend'].value_counts(normalize=True).unstack()
print(siblings_gf_bf)
```
